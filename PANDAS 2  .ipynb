```python
left=pd.DataFrame( {'K':['K0','K2','K5','K3'],
                   'A':['A0','A1','A2','A3'],
                   'B':['B0','B1','B2','B3']
                  })

right=pd.DataFrame( {'K':['K0','K1','K2','K9'],
                   'C':['C0','C1','C2','C3'],
                   'D':['D0','D1','D2','D3']
                  })
```

```python
display(pd.merge(left,right,how = "inner",on=['K']))
```

```python
display(pd.merge(left,right,"inner",'K'))
```

```python
display(pd.merge(left,right,"outer",'K'))
```

```python
display(pd.merge(left,right,"outer",'K'))
display(pd.merge(left,right,"left",'K'))
```

```python
display(pd.merge(left,right,"right",'K'))
```
